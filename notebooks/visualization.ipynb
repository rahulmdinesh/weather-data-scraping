```python
import pandas as pd

# Read the CSV file into a DataFrame with UTF-8 encoding
df = pd.read_csv("../data/corrected_data.csv", encoding='utf-8', sep=';')
df.head(12)
```

**Note:** There's an outlier in 'Precipitation / Rainfall (mm)' , let's analyze that!

```python
import matplotlib.pyplot as plt
```

```python
# Boxplot
plt.figure(figsize=(12, 6))
plt.boxplot(df['Precipitation / Rainfall (mm)'], vert=False)
plt.title('Boxplot of Precipitation / Rainfall (mm)')
plt.xlabel('Total Precipitation / Rainfall (mm)')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(12, 6))
plt.hist(df['Precipitation / Rainfall (mm)'], bins=30)
plt.xlabel('Total Precipitation / Rainfall (mm)')
plt.ylabel('Count')
plt.title('Histogram of Precipitation / Rainfall (mm)')
plt.show()
```

```python
# Display rows where 'Precipitation / Rainfall (mm)' is greater than a certain threshold
high_rainfall = df[df['Precipitation / Rainfall (mm)'] > 1400]

# Select the 'Continent' column
continents = high_rainfall['Continent']

# Count the number of times each continent appears
continent_counts = continents.value_counts()

# Print the counts
print(continent_counts)
```

```python
# Select the 'City' column
cities = high_rainfall['City']

# Print the cities
print(cities)
```

```python
# Group by 'Country' and calculate the sum of 'Precipitation / Rainfall (mm)'
city_rainfall = df.groupby('Continent')['Precipitation / Rainfall (mm)'].sum()

# Sort the results in descending order
city_rainfall = city_rainfall.sort_values(ascending=False)

# Print the results
print(city_rainfall)
```

```python
# Create a larger bar plot
plt.figure(figsize=(20, 10))
city_rainfall.plot(kind='bar')

# Rotate x-axis labels for better readability
plt.xticks(rotation=0)

plt.title('Total Precipitation by Continent')
plt.xlabel('Continent')
plt.ylabel('Total Precipitation / Rainfall (mm)')
plt.show()
```
